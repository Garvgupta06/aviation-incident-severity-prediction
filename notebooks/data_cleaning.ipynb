```python
def clean_injury_severity(x):
    if pd.isna(x):
        return "unknown"
    x = x.strip()

    if x.startswith("Fatal"):
        return "fatal"
    if x == "Incident":
        return "incident"
    if x == "Non-Fatal":
        return "non-fatal"
    return "unknown"

df["injury_severity"] = df["injury_severity"].apply(clean_injury_severity)
    
```

```python
df["aircarft_damage"] = (df["aircraft_damage"].str.lower().str.strip().replace({"unk":"unknown"}))
```

```python
def clean_aircraft_category(x):
    if pd.isna(x):
        return "unknown"
    x = x.strip().lower()

    mapping = {
        "wsft": "weight_shift",
        "ultr": "ultralight"
    }

    # map abbreviations
    if x in mapping:
        return mapping[x]

    # normalize main categories
    x = x.replace("-", "_").replace(" ", "_")

    # rare types → other
    if x in ["rocket", "blimp"]:
        return "other"

    return x

df["aircraft_category"] = df["aircraft_category"].apply(clean_aircraft_category)
```

```python
df["broad_phase_of_flight"] = (
    df["broad_phase_of_flight"]
    .str.lower()
    .str.strip()
    .str.replace("-", "_")
)

df["broad_phase_of_flight"] = df["broad_phase_of_flight"].replace({
    "unknown": "unknown",
    "other": "other"
})
```

```python
df["weather_condition"] = (
    df["weather_condition"]
    .str.upper()
    .str.strip()
    .replace({"UNK": "UNKNOWN"})
    .replace({"UNKNOWN": "unknown", "VMC": "vmc", "IMC": "imc"})
)
```

```python
df["broad_phase_of_flight"] = df["broad_phase_of_flight"].replace({
    "standing": "other"
})
```

```python
df["schedule"] = (
    df["schedule"]
    .str.upper()
    .str.strip()
    .replace({
        "SCHD": "scheduled",
        "NSCH": "nonscheduled",
        "UNK": "unknown"
    })
    .str.lower()
)
```

```python
def clean_purpose(x):
    if pd.isna(x):
        return "unknown"
    x = x.strip().lower()

    # public aircraft mapping
    public_map = ["public aircraft", "public aircraft - federal",
                  "public aircraft - state", "public aircraft - local",
                  "pubs", "publ"]

    if x in public_map:
        return "public_aircraft"

    # aerial operations
    aerial_ops = [
        "aerial application", "aerial observation", "external load",
        "air drop", "glider tow", "firefighting", "skydiving", "banner tow",
        "asho"
    ]
    if x in aerial_ops:
        return "aerial_operations"

    # business
    if x in ["business", "executive/corporate"]:
        return "business"

    # work/movement
    if x in ["positioning", "ferry", "other work use", "flight test", "air race show"]:
        return "other_work_or_movement"

    # base categories
    if x in ["personal", "instructional", "unknown"]:
        return x

    return "other"

df["purpose_of_flight"] = df["purpose_of_flight"].apply(clean_purpose)
```
